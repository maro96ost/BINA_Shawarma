```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_00 = pd.read_csv('teil1.csv')
```

```python
df_25 = pd.read_csv('teil2.csv')
```

```python
df_50 = pd.read_csv('teil3.csv')
```

```python
#merge all 3 imported csv files into one dataframe
df_merged = pd.concat([df_00, df_25, df_50], axis=0)
```

```python
df_merged.count()
```

```python
#drop column "Unnamed: 0"
df_merged.drop('Unnamed: 0',axis = 1, inplace = True)
```

```python
df_merged.head()
```

```python
df_merged
```

```python
#reset index for the merged dataframe
df_merged1 = df_merged.reset_index(drop=True)
```

```python
df = df_merged1
```

```python
#export as a new csv file
df.to_csv('AutoScout_Data.csv', index=False)
```
